```python
import pandas as pd
from statistics import variance
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('var11.csv')
```

```python
data.head()
```

```python
data.describe()
```

Найти среднее и дисперсию каждого столбца.

```python
v1=variance(data['igucj'])
sr1=data['igucj'].mean()
print('Столбец igucj:',v1,sr1)
v2=variance(data['humbk'])
sr2=data['humbk'].mean()
print('Столбец humbk:',v2,sr2)
v3=variance(data['fklbv'])
sr3=data['fklbv'].mean()
print('Столбец fklbv:',v3,sr3)
```

Визуализировать датасет любым удобным способом.

```python
data['igucj'].hist(color="green", label='igucj')
data['humbk'].plot(color="red", label='humbk')
data['fklbv'].hist(color="blue", label='fklbv')
plt.grid(ls = '--', lw = 0.25)
plt.legend()
```

Построить графики для каждой пары признаков в наборе данных 

```python
data['new_col_ig']=(data['igucj']>10).astype('int64')
data['new_col_hu']=(data['humbk']>6).astype('int64')
data['new_col_fk']=(data['fklbv']>10).astype('int64')
data.head()
```

```python
#графикu
fig,axs = plt.subplots(1, 3)
axs[0].plot(data['igucj'])
axs[0].plot(data['humbk'], 'red')
axs[0].grid(ls = '--', lw = 0.25)
axs[0].set_title('new_col_ig_hu')
axs[1].grid(ls = '--', lw = 0.25)
axs[1].set_title('new_col_hu_fk')
axs[1].plot(data['humbk'], 'red')
axs[1].plot(data['fklbv'],'green')
axs[2].plot(data['igucj'])
axs[2].plot(data['fklbv'], 'green')
axs[2].grid(ls = '--', lw = 0.25)
axs[2].set_title('new_col_ig_fk')
fig.tight_layout()
plt.show()
```

Построить гистограмму каждого из признаков.

```python
fig,axs = plt.subplots(1, 3)
axs[0].hist(data['new_col_ig'],data['new_col_ig'].count())
axs[0].set_title('igucj')
axs[1].hist(data['new_col_hu'],data['new_col_hu'].count())
axs[1].set_title('humbk')
axs[2].hist(data['new_col_fk'],data['new_col_fk'].count())
axs[2].set_title('fklbv')
fig.tight_layout()
```

Найти выбросы

```python
data.head()
```

```python
boxplot = data.boxplot(column = ['igucj','humbk','fklbv'])
```

```python
data.head()
```

```python
sr1=data['igucj'].mean()
sr2=data['humbk'].mean()
sr3=data['fklbv'].mean()
print(sr1,sr2,sr3)
plt.plot(data['igucj'],'.-')
# если значение превышает в 10 раз, то это выброс
ig = data['igucj'].apply(lambda x: sr1 if abs(x) > sr1*10 else x)
print(ig)
ig.plot()
```

```python
plt.plot(data['humbk'],'--')
#если значение на 10.1 больше - это выброс
hu= data['humbk'].apply(lambda x: sr2 if abs(x) > sr2+10.1 else x)
hu.plot()
```

```python
plt.plot(data['fklbv'],'--')
#если значение больше на 11 - это выброс
fk = data['fklbv'].apply(lambda x: sr2 if abs(x) > sr2+11 else x)
fk.plot()
```

```python
#Найти новые средние и дисперсии, сравнить со старыми.
news1=ig.mean()
news2=hu.mean()
news3=fk.mean()
print(news1-sr1,news2-sr2,news3-sr3)

newv1=variance(ig)
newv2=variance(hu)
newv3=variance(fk)
print(newv1-v1,newv2-v2,newv3-v3)
```

```python
#Нормализовать данные по среднему и дисперсии.
fig, axs = plt.subplots(1,3,figsize=(20,5))
axs[0].plot(ig.apply(lambda x: (x - news1)/newv1),'b.-')
axs[0].set_title('fubjw')
axs[0].grid(ls = '--', lw = 0.25)
axs[1].plot(hu.apply(lambda x: (x - news2)/newv2),'r.-')
axs[1].set_title('pwjpx')
axs[1].grid(ls = '--', lw = 0.25)
axs[2].plot(fk.apply(lambda x: (x - news3)/newv3),'g.-')
axs[2].set_title('yxgod')
axs[2].grid(ls = '--', lw = 0.25)
fig.tight_layout()
plt.show()
```
